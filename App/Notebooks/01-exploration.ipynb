# Contexto

Este conjunto de datos incluye información para la estimación de niveles de obesidad en individuos de los países de México, Perú y Colombia, basada en sus hábitos alimenticios y condición física. El 23% de los datos se recolectó directamente de los usuarios a través de una encuesta realizada por Fabio Mendoza Palechor y Alexis de la Hoz Manotas en una plataforma web y el 77% de los datos se generó sintéticamente utilizando la herramienta Weka y el filtro SMOTE.

El conjunto de datos contiene 19 atributos y 2111 registros.

- Género es 1 si un encuestado es masculino y 0 si es femenino.
- Edad es la edad de un encuestado en años.
- family_history_with_overweight es 1 si un encuestado tiene un familiar que es o fue con sobrepeso, 0 si no.
- FAVC es 1 si un encuestado consume alimentos altos en calorías frecuentemente, 0 si no.
- FCVC es 1 si un encuestado suele comer vegetales en sus comidas, 0 si no.
- NCP representa cuántas comidas principales tiene un encuestado diariamente (0 para 1-2 comidas, 1 para 3 comidas y 2 para más de 3 comidas).
- CAEC representa cuánta comida consume un encuestado entre comidas en una escala de 0 a 3.
- SMOKE es 1 si un encuestado fuma, 0 si no.
- CH2O representa cuánta agua bebe un encuestado en una escala de 0 a 2.
- SCC es 1 si un encuestado monitorea su ingesta calórica, 0 si no.
- FAF representa cuánta actividad física realiza un encuestado en una escala de 0 a 3.
- TUE representa cuánto tiempo pasa un encuestado mirando dispositivos con pantallas en una escala de 0 a 2.
- CALC representa con qué frecuencia un encuestado consume alcohol en una escala de 0 a 3.
- Automóvil, Bicicleta, Motocicleta, Transporte Público y Caminar indican el modo principal de transporte de un encuestado. Su modo principal de transporte se indica con un 1 y las otras columnas contendrán un 0.
- NObeyesdad es 1 si un paciente es obeso y 0 si no.
